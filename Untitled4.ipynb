## Predictive modeling

### Subtask:
User Segmentation with Clustering: Use clustering algorithms (e.g., K-Means) to segment users into distinct groups based on their behavior and characteristics.


**Reasoning**:
Select relevant features, standardize them, and apply K-Means clustering. Then, analyze the clusters by examining the mean values of the original features within each cluster.



```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Select features for clustering (excluding Churn, User_ID, and User_ID encoded columns)
clustering_features = df_encoded.drop(columns=[col for col in df_encoded.columns if 'User_ID' in col or col == 'Churn'])

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(clustering_features)

# Determine the optimal number of clusters using the Elbow method
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()

# Based on the elbow plot, let's choose k=3 (or another visually determined elbow point)
optimal_k = 3

# Initialize and fit the K-Means model with the chosen number of clusters
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(scaled_features)

# Analyze the characteristics of each cluster
cluster_analysis = df.groupby('Cluster').agg({
    'User_ID': 'count',
    'App_Installed': 'mean',
    'First_Login_Completed': 'mean',
    'Days_Since_Last_Activity': 'mean',
    'Registered_for_Event': 'mean',
    'Newsletter_Subscribed': 'mean',
    'Saved_Event_Count': 'mean',
    'Feedback_Rating': 'mean',
    'Course_Completed': 'mean',
    'Time_Spent_Total_Minutes': 'mean',
    'user_loyalty': 'mean',
    'Churn': 'mean'
}).reset_index()

display(cluster_analysis)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Load the dataset from the provided URL
url = "https://docs.google.com/spreadsheets/d/1TfjvrR0nkTIWF3GIzorW4BUe4dR0MkgO/export?format=csv"
df = pd.read_csv(url)

# Identify categorical columns
categorical_cols = df.select_dtypes(include='object').columns

# Apply one-hot encoding
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Select features for clustering
clustering_features = df_encoded.drop(columns=[col for col in df_encoded.columns if 'User_ID' in col])

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(clustering_features)

# Apply K-Means clustering
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(scaled_features)

# Save the dataframe with the cluster information
df.to_csv('clustered_dataset.csv', index=False)

print("The dataset with the clustering information has been saved to 'clustered_dataset.csv'")
```
